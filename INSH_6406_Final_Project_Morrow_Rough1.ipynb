```python
import pandas
```

```python
bills = pandas.read_csv('data/CA1718/bills.csv', encoding = 'utf-8', sep = ',')
bills
```

```python
people = pandas.read_csv('data/CA1718/people.csv', encoding = 'utf-8', sep = ',')
people
```

```python
sponsors = pandas.read_csv('data/CA1718/sponsors.csv', encoding = 'utf-8', sep = ',')
with open('data/CA1718/sponsors.csv', 'r', encoding = 'utf-8') as f:
    header = f.readline().strip().split(',')
    sponsorslist = [line.strip().split(',') for line in f.readlines()]
sponsors
sponsorslist
```

```python
# Creates a dictionary of people where Name = Key, people_id = Value, converts it to a list and a pandas dataframe
peopledict = dict(zip(people['name'], people['people_id']))
peoplelist = []
for key, value in peopledict.items():
    newlist = key,value
    peoplelist.append(newlist)
peopledf = pandas.DataFrame(peoplelist, columns=('name','pid'))
peoplelist
```

```python
# Matches sponsors with bills
for items in list(sponsors['people_id']):
    for name, pid in peoplelist:
        if int(items) == int(pid):
            sponsors['sponsor'] = name
sponsors
```

```python
sponsorslist
for items in sponsorslist:
    for name, pid in peoplelist:
        if int(items[1]) == int(pid):
            items.append(name)
sponsoredlist = pandas.DataFrame(sponsorslist, columns=('bill','pid','position','sponsor'))
sponsoredlist
```
